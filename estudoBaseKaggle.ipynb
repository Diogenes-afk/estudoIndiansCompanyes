```python
semDup = df.drop_duplicates(subset=['company_name', 'rating'])
semDup
```

```python
##tratando a coluna reviews
rev=[]
for i in df['company_reviews']:
    a=i.split(sep='k')[0][1:]
    rev.append(float(a)*1000)
rev
```

```python
##tratando a coluna idade
age=[]
for i in df['company_age']:
    a=i.split(sep=' ')[0]
    age.append(a)
age
```

```python
df['company_reviews'] = pd.Series(rev)
df['company_age'] = pd.Series(age).astype(int)
```

```python
sns.set_color_codes("muted")
sns.barplot(x="company_reviews", y="company_name", data=df,
            label="total", color="b")
```
